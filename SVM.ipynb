```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("C:/ML Course/Datasets/placement_data.csv")
```

```python
dataset.head(5)
```

```python
sns.scatterplot(x="cgpa",y="score", data=dataset, hue="selection")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["selection"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel = "poly")
sv.fit(x_train, y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = sv)
plt.show()
```

```python
sv.predict([[10,11]])
```

# Regression 

```python
data = pd.read_excel("package_prediction.xlsx")
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x="CGPA",y="PACKAGES",data=data)
plt.show()
```

```python
x = data[["CGPA"]]
y = data["PACKAGES"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel="linear")
svr.fit(x_train,y_train)
```

```python
svr.score(x_test,y_test)
```

```python
svr.score(x_train, y_train)
```

```python
sns.scatterplot(x="CGPA",y="PACKAGES",data=data)
plt.plot(data["CGPA"],svr.predict(x),color="red")
plt.show()
```
