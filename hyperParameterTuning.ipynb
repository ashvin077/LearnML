```python
import pandas as pd
```

```python
dataset = pd.read_excel(r"C:\ML Course\Datasets\level_salary.xlsx")
```

```python
dataset
```

```python
x = dataset[["level"]]
y = dataset["salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
dt.score(x_test, y_test)
```

```python
dt.score(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
df={"criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter":["best", "random"],
    "max_depth":[i for i in range(2,20)]}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=df)
```

```python
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
dt1 = DecisionTreeRegressor(criterion='squared_error',splitter='best',max_depth=8)
```

```python
dt1.fit(x_train,y_train)
```

```python
dt1.score(x_train,y_train), dt1.score(x_test,y_test)
```

```python
gd.best_score_
```

# RandomizeSearchCV

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
```

```python
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_
```
