```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\ML Course\Datasets\placement_data.csv")
```

```python
dataset.head(5)
```

```python
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="selection")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["selection"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=48)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
```

```python
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100, pr.score(x_test,y_test)*100
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)*100, knn.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```
