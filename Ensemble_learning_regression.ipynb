```python
import pandas as pd
```

```python
dataset = pd.read_excel("package_prediction.xlsx")
```

```python
dataset.head(3)
```

```python
x = dataset[["CGPA"]]
y = dataset["PACKAGES"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```python
svm = SVR()
svm.fit(x_train,y_train)
svm.score(x_train,y_train)*100, svm.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr1",LinearRegression()), ("dt",DecisionTreeRegressor()),("svm",SVR())]
```

```python
vc = VotingRegressor(li,weights=[10,20,30])
```

```python
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train), vc.score(x_test,y_test)
```

```python
df = {
    "lr":lr.predict(x_test),
    "dt":dt.predict(x_test),
    "svm":svm.predict(x_test),
    "vc":vc.predict(x_test)
}
```
