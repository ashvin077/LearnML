```python
import pandas as pd
```

```python
from sklearn import datasets
```

```python
dataset = pd.read_csv("IRIS.csv")
dataset
```

```python
dataset.columns
```

```python
dataset['species'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dataset['species'] = le.fit_transform(dataset['species'])
```

```python
dataset
```

```python
data = pd.DataFrame({
            'sepal_length': dataset.loc[:,'sepal_length'],
            'sepal_width': dataset.loc[:,'sepal_width'],
            'petal_length': dataset.loc[:,'petal_length'],
            'petal_width': dataset.loc[:,'petal_width'],
            'species': dataset.loc[:,'species']
             })
data
```

```python
x = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data['species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test)
```

```python
x_test            # 0: setosa, 1: versicolor, 2: virginica
```

```python
clf.score(x_test,y_test)*100
```
